1 create first chart


```python
import matplotlib.pyplot as plt # visualization
x = [2010,2015,2020,2025] # x cord
y = [100,200,250,300] # y cord
plt.plot(x,y) # line graph
plt.xlabel("years") # x label
plt.ylabel("sales") # y label
plt.title("sales report") # graph title
plt.show() # graph show
```

```python
import matplotlib.pyplot as plt # visualization
x = [2010,2015,2020,2025] # x cord
y = [100,200,250,300] # y cord

# graph size
plt.figure(figsize=(6,4)) # width or height
plt.plot(x,y)
plt.show()
```

2 customized chart

```python
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y = [100,200,250,300]
plt.plot(x,y,color="y",marker="*",linestyle=":",linewidth=4,markersize=14)
plt.show()
```

3 multiple line chart

```python
import matplotlib.pyplot as plt

x =[2010,2015,2020,2025]
y1 = [100,200,260,290]
y2 = [150,185,195,300]

plt.plot(x,y1,label="jeans")
plt.plot(x,y2,label="shirt")
plt.legend() # info of label
plt.show()
```

```python
import matplotlib.pyplot as plt

x =[2010,2015,2020,2025]
y1 = [100,200,260,290]
y2 = [150,300,350,500]
y3 = [300,350,400,450]

plt.plot(x,y1,label="jeans",color="g",marker="*",linestyle=":")
plt.plot(x,y2,label="shirt",color="y",marker="o",linestyle="dashed")
plt.plot(x,y3,label="shoes",color="b",marker="*",linestyle="--")
plt.legend() # info of label
plt.show()

```

3 Bar chart

```python
import matplotlib.pyplot as plt
x = [2015,2020,2025,2030]
y = [100,150,200,290]

plt.bar(x,y)
# size
plt.figure(figsize=(6,2))
plt.show()
```

```python
# multiple bar chart
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y1 = [10,20,20,40]
y2 = [20,30,25,30]
# calculation -> width
w = 0.40

plt.bar(x - w/2,y1,label="boys",width=w)
plt.bar(x + w/2,y2,label="girls",width=w)
plt.xlabel("groups")
plt.ylabel("number of students")
plt.title("number of students in each group")
plt.legend()
plt.show()
```
